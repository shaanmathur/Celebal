# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"

````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
def agent(query: str):
    query_lower = query.lower()

    try:
        if "calculate" in query_lower:
            expression = query_lower.replace("calculate", "").strip()
            result = calculator(expression)
            return {"type": "calculation", "result": result}

        elif "keywords" in query_lower:
            result = extract_keywords(query)
            return {"type": "keywords", "result": result}

        else:
            return {
                "type": "general",
                "result": "I can help with calculations or keyword extraction."
            }

    except Exception as e:
        return {"type": "error", "result": str(e)}
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````

````python
import datetime

# 🛠️ TOOL 3: Word Counter (bonus tool)
def word_counter(text: str) -> int:
    """Count the number of words in a text."""
    try:
        return len(text.split())
    except Exception:
        return 0


def log_event(query, response):
    """Simple logging of each agent call."""
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"[LOG {timestamp}] Query: '{query}' -> Type: {response['type']}")


def agent(query: str):
    query_lower = query.lower()

    try:
        if "calculate" in query_lower:
            expression = query_lower.replace("calculate", "").strip()
            result = calculator(expression)
            response = {"type": "calculation", "result": result}

        elif "keywords" in query_lower:
            result = extract_keywords(query)
            response = {"type": "keywords", "result": result}

        elif "count words" in query_lower or "word count" in query_lower:
            text = query_lower.replace("count words", "").replace("word count", "").strip()
            result = word_counter(text)
            response = {"type": "word_count", "result": result}

        else:
            response = {
                "type": "general",
                "result": "I can help with calculations, keyword extraction, or word counting."
            }

    except Exception as e:
        response = {"type": "error", "result": str(e)}

    log_event(query, response)
    return response
````

````python
queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "Count words in this is a simple test sentence",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````
